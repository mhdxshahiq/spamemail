```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
raw_data = pd.read_csv(r'mail_data.csv')
```

```python
raw_data.head()
```

```python
raw_data.isnull().sum()
```

```python
raw_data.shape
```

```python
raw_data.loc[raw_data['Category'] == 'spam', 'Category'] = 0
raw_data.loc[raw_data['Category'] == 'ham', 'Category'] = 1
```

```python
x = raw_data['Message']
y = raw_data['Category']
```

```python
print(y)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=3)
```

```python
featurizer = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)
xtrain_features = featurizer.fit_transform(xtrain)
xtest_features = featurizer.transform(xtest)

ytrain = ytrain.astype('int')
ytest = ytest.astype('int') 
```

```python
print(xtrain)
```

```python
print(xtest_features)
```

```python
model = LogisticRegression()
```

```python
model.fit(xtrain_features, ytrain)
```

```python
predictions = model.predict(xtrain_features)
accuracy_score = accuracy_score(ytrain, predictions)
print("Accuracy score on training data : ", accuracy_score)
```

```python
y_pred = model.predict(xtest_features)
accuracy = accuracy_score(ytest, y_pred)

print("Accuracy score on test data:", accuracy)
```

```python
input_mail = ["Congratulations! You've won a $1,000 Walmart gift card. Click here to claim your prize."]
input_mail_features = featurizer.transform(input_mail)
prediction = model.predict(input_mail_features)
print(prediction)  # Output: [0] for spam, [1] for ham
```
